## 11. Contract

The column `Contract` displays the players contract details
> The goal is to create 3 new columns that will display the start and end year of the players contract
> for all players who are not on loan or free and to calculate the length f their contract from these
> informations.

```python
fifa['Contract'].unique()
```

```python
for index, row in fifa.iterrows():
    if 'On Loan' in row['Contract'] or 'Free' in row['Contract']:
        print(row['Contract'])
```

```python
def extract_contract_info(contract):
    if contract == 'Free' or 'On Loan' in contract:
        starting_date = np.nan
        end_date = np.nan
        contract_length = 0
    else:
        starting_date, end_date = contract.split(' ~ ')
        starting_year = int(starting_date[:4])
        end_year = int(end_date[:4])
        contract_length = int(end_year - starting_year)
    return starting_date, end_date, int(contract_length)

# Apply this to contract column and get the new columns in
new_cols = ['Contract Start', 'Contract End', 'Contract Length(years)']
new_data = fifa['Contract'].apply(lambda x: pd.Series(extract_contract_info(x)))

for i in range(len(new_cols)):
    fifa.insert(loc=fifa.columns.get_loc('Contract')+1+i, column = new_cols[i], value = new_data[i])
```

```python
fifa[['Contract', 'Contract Start', 'Contract End', 'Contract Length(years)']].sample(3)
```

```python
# adding a contract categorization for better data handling
def contract_categorizing(contract):
    if contract == 'Free':
        return 'Free'
    elif 'On Loan' in contract:
        return 'On Loan'
    else:
        return 'Contract'

# adding contract_status column
fifa.insert(fifa.columns.get_loc('Contract Length(years)')+1, 'Contract Status', fifa['Contract'].apply(contract_categorizing))
pd.set_option('display.max_columns', None) 
fifa.sample(3)
```

## 13. Height

The column `Height` displays the player's height.

```python
fifa['Height'].unique()
```

### We need a function to convert the heights from ft to cmüìè

```python
def convert_heights(height):
    if 'cm' in height:
        return int(height.strip('cm'))
    else:
        feet, inches = height.split("'")
        total_inches = int(feet)*12 + int(inches.strip('"'))
        return round(total_inches * 2.54)
# apply to height col
fifa['Height'] = fifa['Height'].apply(convert_heights)
fifa['Height'].unique()
```

```python
fifa = fifa.rename(columns = {'Height': 'Height in cm'})
fifa.sample(3)
```

## 14. Weight

### This column displays the players weight, the problem is that some are in lb and most of them in kg!üèãÔ∏è‚Äç‚ôÇÔ∏è

```python
fifa['Weight'].unique()
```

```python
# weight converter function
def convert_weight(weight):
    if 'kg' in weight:
        return int(weight.strip('kg'))
    else:
        pounds = int(weight.strip('lbs'))
        return round(pounds / 2.205) 

# apply the function
fifa['Weight'] = fifa['Weight'].apply(convert_weight)
fifa['Weight'].unique()
```

```python
fifa = fifa.rename(columns = {'Weight': 'Weight in kg'})
fifa.sample(3)
```

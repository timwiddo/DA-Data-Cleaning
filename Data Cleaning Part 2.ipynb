Now we want to split up the span

```python
df['Span'].str.split(pat = '-')
```

```python
df['Rookie_Year'] = df['Span'].str.split(pat = '-').str[0]
df['Final_Year'] = df['Span'].str.split(pat = '-').str[1]
df.head()
```

```python
# trying to calculate the span in years
for x in df.index:
    start_year, end_year = df['Span'][x].split('-')
    span_years = int(end_year) - int(start_year)
    if span_years > 1:
         df.loc[x, 'Span'] = str(span_years) + ' yrs'
    else:
         df.loc[x, 'Span'] = str(span_years) + ' yr'
    
df.rename(columns={'Span': 'CareerLength'}, inplace=True)
df.head()
```

Split the country from the Player


```python
df['Player'].str.split(pat ='(')
```

```python
df['Country'] = df['Player'].str.split(pat = '(').str[1]
df['Country']
```

```python
df['Country'] = df['Country'].str.replace(')','')
df['Country']
```

```python
df['Player'] = df['Player'].str.split(pat = '(').str[0]
df.head()
```

```python
df.dtypes
```

```python
# get rid of * in a more advanced way, replace wouldve also worked
df['Highest_Ins_Score'] = df['Highest_Ins_Score'].str.split(pat = '*').str[0]
df
```

```python
df['Highest_Ins_Score'] = df['Highest_Ins_Score'].astype('int')
df.dtypes
```

```python
#map typechanges
df = df.astype({'Rookie_Year': 'int',
                'Final_Year': 'int',
                'Mat': 'int'})
df.dtypes
```

### 1. What is the Mean Career length?

```python
test = df['CareerLength'].str.split(' ').str[0]
test = pd.to_numeric(test, errors='coerce')
test.mean()
```

### 2. What is the average batting strike rate for player who have played for over 10 years?

```python
filtered_df = df[df['CareerLength'].str.split().str[0].astype(int) > 10]
filtered_df['Batting_Strike_Rate'].mean()
```

### 3. What is the Number of Players who started playing before 1960?

```python
Result = df[df['Rookie_Year'] < 1960]['Player'].count()
Result
```

### 4. What is the highest inning score per Country?

```python
df_grouped = df.groupby('Country')['Highest_Ins_Score'].max().to_frame('Highest_Inning_Scores_by_Country')
df_grouped
```

```python
df_grouped.reset_index().sort_values('Highest_Inning_Scores_by_Country', ascending = False)
```
